# 2)-Loading data

```python
data=pd.read_excel('data_10000_eng.xlsx')
data.shape
```

```python
data.head()
```

```python
list_language = ['USSD','UKSD']
```

```python
l=data[['title','firstmessage','dep','firstusedtextblock','language','ticketnumber']].astype(str) # this converts all values inside into string
```

```python
for language in list_language:
    l[language] = l.astype(str).sum(axis=1).str.contains(language)
```

```python
l.shape
```

```python
l.head()
```

```python
l.USSD.value_counts()
```

# 3- Creating English Dataset

```python
us_df=l.loc[l.ticketnumber.str.contains('USSD',case=False)]
us_df
```

```python
uk_df=l.loc[l.ticketnumber.str.contains('UKSD',case=False)]
uk_df
```

```python
df_eng = pd.concat([us_df, uk_df])

df_eng
```
